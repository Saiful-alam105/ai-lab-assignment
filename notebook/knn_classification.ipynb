```python
import pandas as pd
import random
import math
```

```python
iris = pd.read_csv("../dataset/iris.csv")
```

```python
iris.head()
```

```python
print("Row, Column:", iris.shape)
```

```python
dataset = iris.values.tolist()

for row in dataset[:5]:
    print(row)
```

```python
#Dataset Split Function

def split_dataset(dataset):

    train_set = []
    val_set = []
    test_set = []
    for sample in dataset:
        R = random.random()
        if R <= 0.70:
            train_set.append(sample)
        elif R <= 0.85:
            val_set.append(sample)
        else:
            test_set.append(sample)

    return train_set, val_set, test_set

train_set, val_set, test_set = split_dataset(dataset)

print("Train set:", len(train_set))
print("Validation set:", len(val_set))
print("Test set:", len(test_set))
```

```python
# Euclidean Distance Function

def euclidean_distance(sample1, sample2):

    distance = 0

    # exclude last column (class label)
    for i in range(len(sample1)-1):

        distance += (sample1[i] - sample2[i])**2

    return math.sqrt(distance)

print(euclidean_distance(train_set[0], train_set[1]))
```

```python
# KNN Classification Function

def knn_classification(train_set, validation_sample, K):

    distances = []

    # calculate distance from validation sample to every training sample
    for training_sample in train_set:

        distance = euclidean_distance(validation_sample, training_sample)

        # store distance + class label
        distances.append((distance, training_sample[-1]))

    # sort by smallest distance
    distances.sort()

    # take K nearest neighbors
    neighbors = distances[:K]

    class_count = {}

    # majority voting
    for neighbor in neighbors:

        label = neighbor[1]

        if label not in class_count:
            class_count[label] = 0

        class_count[label] += 1

    # choose most common class
    predicted_class = max(class_count, key=class_count.get)

    return predicted_class
```

```python
# Test KNN Function

prediction = knn_classification(train_set,val_set[0],5)

print("Prediction:", prediction)
print("Actual:", val_set[0][-1])
```

```python
# Accuracy Function

def calculate_accuracy(train_set, validation_set, K):

    correct = 0

    for sample in validation_set:

        predicted = knn_classification(train_set,sample,K)

        actual = sample[-1]

        if predicted == actual:
            correct += 1

    accuracy = (correct / len(validation_set)) * 100

    return accuracy
```

```python
# Validation Experiment

K_values = [1,3,5,10,15]

results = []

for K in K_values:

    accuracy = calculate_accuracy(train_set,val_set,K)

    results.append([K, accuracy])

    print("K =", K,"| Accuracy =", round(accuracy,2))

```

```python
# Create Results Table

results_df = pd.DataFrame(results,columns=["K","Validation Accuracy"])

results_df
```

```python
# Find Best K

best_row = max(results,key=lambda x: x[1])

best_K = best_row[0]
best_accuracy = best_row[1]

print("Best K:", best_K)
print("Best Validation Accuracy:",round(best_accuracy,2))
```

```python
# Final Test Accuracy

test_accuracy = calculate_accuracy(train_set,test_set,best_K)

print("Final Test Accuracy:",round(test_accuracy,2))
```
